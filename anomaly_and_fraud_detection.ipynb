```python
import pandas as pd
df=pd.read_csv('anomaly_detection_results.csv')
transactions=pd.read_csv('normalise_train.csv')
```

```python
transactions = transactions.copy()  # Ensure we don't modify the original in place
transactions['Is_Anomalous'] = df.loc[transactions.index, 'Is_Anomalous'].values
transactions['Reconstruction_Error'] = df.loc[transactions.index, 'Reconstruction_Error'].values
transactions
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score

# Step 1: Extract the `Is_Anomalous` and `label` columns from the `transactions` DataFrame
y_true = transactions['label']  # Ground truth labels
y_pred = transactions['Is_Anomalous'].astype(int)  # Predicted labels from the autoencoder

# Step 2: Calculate the confusion matrix
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

# Step 3: Calculate the scores
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

# Step 4: Display the results
print("Confusion Matrix (TN, FP, FN, TP):", (tn, fp, fn, tp))
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
# we need to increase the precision
```

```python
threshold=0.000636
df=transactions[transactions['Reconstruction_Error']>threshold]
```

```python
import numpy as np

# Calculate thresholds
mean_velocity = df['transactions_per_day'].mean()
std_velocity = df['transactions_per_day'].std()
high_velocity_threshold = mean_velocity + 2 * std_velocity  # Adjust the multiplier for sensitivity

# Detect high velocity anomalies
df['High_Velocity_Flag'] = df['transactions_per_day'] > high_velocity_threshold
df[df['High_Velocity_Flag']==True].size
```

```python
# Detect odd-hour transactions
df['Odd_Hour_Flag'] = df['hour'].apply(lambda x: 1 if x < 4 or x > 23 else 0)

# Aggregate odd-hour flags per merchant
odd_hour_counts = df.groupby('merchant_id')['Odd_Hour_Flag'].sum().reset_index()
odd_hour_counts.rename(columns={'Odd_Hour_Flag': 'Odd_Hour_Transactions'}, inplace=True)

# Merge back into the main dataset (optional)
df = df.merge(odd_hour_counts, on='merchant_id', how='left')
odd_hour_counts.sum()
```

```python
# Calculate customer concentration ratio
total_customers = df.groupby('merchant_id')['customer_id'].nunique()
total_transactions = df.groupby('merchant_id')['transaction_id'].count()
customer_concentration_ratio = (total_customers / total_transactions).reset_index()

# Rename columns for clarity
customer_concentration_ratio.rename(
    columns={0: 'Concentration_ratio'}, inplace=True
)

# Merge back into the main dataset (optional)
df = df.merge(customer_concentration_ratio, on='merchant_id', how='left')

# Flag high concentration merchants (custom threshold, e.g., < 0.2)
df['High_Concentration_Flag'] = df['Concentration_ratio'] < 0.2
df[df['High_Concentration_Flag']==True].size
```

```python
# Assign weights to flags
weights = {
    'High_Velocity_Flag': 0.5,
    'Odd_Hour_Flag': 0.3,
    'High_Concentration_Flag': 0.2
}

# Calculate fraud score per transaction
df['Fraud_Score'] = (
    weights['High_Velocity_Flag'] * df['High_Velocity_Flag'].astype(int) +
    weights['Odd_Hour_Flag'] * df['Odd_Hour_Flag'].astype(int) +
    weights['High_Concentration_Flag'] * df['High_Concentration_Flag'].astype(int)
)

# Flag anomalies based on fraud score threshold
# fraud_score_threshold = 0.5  # Adjust threshold as needed
# df['Is_Anomalous'] = df['Fraud_Score'] > fraud_score_threshold
```

```python
# Filter for each flag and calculate mean and std fraud score
mean_fraud_score_high_velocity = df[df['High_Velocity_Flag'] == 1]['Fraud_Score'].mean()
std_fraud_score_high_velocity = df[df['High_Velocity_Flag'] == 1]['Fraud_Score'].std()

mean_fraud_score_odd_hour = df[df['Odd_Hour_Flag'] == 1]['Fraud_Score'].mean()
std_fraud_score_odd_hour = df[df['Odd_Hour_Flag'] == 1]['Fraud_Score'].std()

mean_fraud_score_customer_concentration = df[df['High_Concentration_Flag'] == 1]['Fraud_Score'].mean()
std_fraud_score_customer_concentration = df[df['High_Concentration_Flag'] == 1]['Fraud_Score'].std()

# Print the results
print("Fraud Scores (Mean and Std) for Different Rules:")
print(f"High Velocity Detection - Mean: {mean_fraud_score_high_velocity}, Std: {std_fraud_score_high_velocity}")
print(f"Odd-Hour Pattern Detection - Mean: {mean_fraud_score_odd_hour}, Std: {std_fraud_score_odd_hour}")
print(f"Customer Concentration Analysis - Mean: {mean_fraud_score_customer_concentration}, Std: {std_fraud_score_customer_concentration}")
```
